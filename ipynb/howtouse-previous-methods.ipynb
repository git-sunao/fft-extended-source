```python
# set path
import sys, os
sys.path.append(os.path.dirname(os.getcwd()))

import numpy as np
import matplotlib.pyplot as plt
import python
import mlfast
```

## FFT-based method

```python
mag_disk = mlfast.mag_fft.magnification_disk()
mag_limb1= mlfast.mag_fft.magnification_limb(1)
mag_limb2= mlfast.mag_fft.magnification_limb(2)
```

## Witt Mao analytic

```python
u = np.linspace(0.0, 4.0 ,1000)
rho = 0.2
%time a_disk = mag_disk.A(u, rho)
%time a_disk_WittMao1994 = python.mag_wm.A_disk(u, rho)

plt.figure()
plt.yscale('log')
plt.xlabel(r'$u$')
plt.ylabel(r'$A(u)$')
plt.plot(u, a_disk)
plt.plot(u, a_disk_WittMao1994, ls='--')
plt.show()
```

```python
u = np.linspace(0.0, 4.0 ,1000)
rho = 0.2
%time a_disk = mag_disk.A(u, rho)
%time a_disk_Gould = python.mag_yoo.A_disk(u, rho)

plt.figure()
plt.yscale('log')
plt.xlabel(r'$u$')
plt.ylabel(r'$A(u)$')
plt.plot(u, a_disk)
plt.plot(u, a_disk_Gould, ls='--')
plt.show()
```
